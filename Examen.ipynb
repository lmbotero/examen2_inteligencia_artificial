```python
import random
from datetime import datetime

poblacion_inicial = " abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ"
string_final = "Maria Fernanda Sanchez Cardona"
     
def generar_poblacion(length, poblacion):
    poblacion_generada = []
    while len(poblacion_generada) < length:
        tamano = min(length - len(poblacion_generada), len(poblacion))
        poblacion_generada.extend(random.sample(poblacion, tamano))
    return ''.join(poblacion_generada)

def obtener_fitness(poblacion_generada, string_final):
    return sum(1 for expected, actual in zip(string_final, poblacion_generada)
           if expected == actual)

def mutar(padre, poblacion_inicial):
    index = random.randrange(0, len(padre))
    hijos = list(padre)
    nuevosHijos, alternate = random.sample(poblacion_inicial, 2)
    hijos[index] = alternate \
        if nuevosHijos == hijos[index] \
        else nuevosHijos
    return ''.join(hijos)

def imprimir(hijo, fitness, tiempo_inicial, generaciones, probabilidad):
          diferencia_tiempo = datetime.now() - tiempo_inicial
          print("{0}\t\t{1}\t{2}\t\t{3}\t\t{4}".format(str(generaciones), fitness, hijo, str(diferencia_tiempo), str(probabilidad)))

def calcular_probabilidad(fitness, totalFitness):
   probabilidad = 0
   if totalFitness > 0:
      probabilidad = fitness / totalFitness
   return probabilidad

totalFitness = 0
generaciones = 0
random.seed(12345)
tiempo_inicio = datetime.now()
mejor_padre = generar_poblacion(len(string_final), poblacion_inicial)
mejor_fitness = obtener_fitness(mejor_padre, string_final)
totalFitness = totalFitness + mejor_fitness
print("{0}\t{1}\n".format("Población inicial: ", str(poblacion_inicial)))
print("{0}\t{1}\n".format("Población final: ", str(string_final)))
print("Comienza el algoritmo genético")
timeDiff = datetime.now() - tiempo_inicio
print("{0}\t{1}\n".format(mejor_padre, str(timeDiff)))

print("{0}\t{1}\t{2}\t{3}\t\t\t{4}\n".format("# Generación", "Fitness", "Solución", "Tiempo", "Probabilidad"))

for i in range(20000):
    generaciones = generaciones + 1
    child = mutar(mejor_padre, poblacion_inicial)
    child_fitness = obtener_fitness(child, string_final)
    probabilidadHijo = calcular_probabilidad(child_fitness, totalFitness)
    if child_fitness >= mejor_fitness:
        imprimir(child,child_fitness, tiempo_inicio, generaciones, probabilidadHijo)
        totalFitness = totalFitness + child_fitness
        mejor_fitness = child_fitness
        mejor_padre = child

    if child_fitness >= len(mejor_padre):
        print("\nResultado final:\n")
        probabilidadHijo = calcular_probabilidad(child_fitness, totalFitness)
        imprimir(child,child_fitness, tiempo_inicio, generaciones, probabilidadHijo)
        break

print("\n")
print("{0}{1}\n".format("Número de generaciones: ", str(generaciones)))
print("Termina el algoritmo genético")
```
